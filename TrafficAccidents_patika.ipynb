Columns that have mostly missing values are dropped.

```python
df.drop(columns=[
    'Junction_Detail',
    'Carriageway_Hazards',
    'Special_Conditions_at_Site',
    'Junction_Control'
], inplace=True, errors='ignore')
```

## Visualization

Yearly accident count is calculated.

```python
df['Year'].value_counts().sort_index().plot(kind='bar')
plt.title('Accident Count per Year')
plt.xlabel('Year')
plt.ylabel('Accident Count')
plt.show()
```

Distribution of accidents based on weekdays is investigated.

```python
day_map = {
    1: 'Sunday', 2: 'Monday', 3: 'Tuesday',
    4: 'Wednesday', 5: 'Thursday', 6: 'Friday', 7: 'Saturday'
}

df['Day_of_Week'] = df['Day_of_Week'].map(day_map)

# Order the days explicitly
ordered_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_counts = df['Day_of_Week'].value_counts().reindex(ordered_days)

# Plot pie chart
colors = sns.color_palette("Set2")

plt.figure(figsize=(8, 8))
plt.pie(
    day_counts,
    labels=ordered_days,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors
)
plt.title('Distribution of Accidents by Week Day')
plt.axis('equal')  # Equal aspect ratio ensures pie is drawn as a circle
plt.show()
```

Accidents appear to be distributed mostly equally throughout the week.

Hourly distribution of accidents is investigated.

```python
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M').dt.hour

hour_counts = df['Hour'].value_counts().sort_index()

colors = sns.color_palette("Set2", n_colors=len(hour_counts))

plt.figure(figsize=(10, 5))
plt.bar(hour_counts.index, hour_counts.values, color=colors)
plt.title('Number of Accidents by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Accidents')
plt.xticks(range(0, 24))
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

As expected, most accidents accumulate during work hours.

Distribution of accidents based on light conditions are investigated.

```python
light_counts = df['Light_Conditions'].value_counts()

colors = sns.color_palette("Set2", n_colors=len(light_counts))

plt.figure(figsize=(10, 5))
plt.bar(light_counts.index, light_counts.values, color=colors)
plt.title('Number of Accidents by Light Conditions')
plt.xlabel('Light Condition')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

Most accidents occur during day light.

Distribution of casualties is investigated.

```python
casualty_counts = df['Number_of_Casualties'].value_counts().sort_index()

# Generate a color palette with the right number of colors
colors = sns.color_palette("Set2", n_colors=len(casualty_counts))

# Plot using matplotlib to apply colors
plt.figure(figsize=(10, 5))
plt.bar(casualty_counts.index.astype(str), casualty_counts.values, color=colors)
plt.title('Distribution of Accidents by Number of Casualties')
plt.xlabel('Number of Casualties')
plt.ylabel('Number of Accidents')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

Majority of accidents have 1 casualy. Multiple casualties are rare.

```python
weather_counts = df['Weather_Conditions'].value_counts()

# Generate color palette with matching number of colors
colors = sns.color_palette("Set2", n_colors=len(weather_counts))

# Plot using matplotlib to control colors
plt.figure(figsize=(10, 5))
plt.bar(weather_counts.index, weather_counts.values, color=colors)
plt.title('Number of Accidents by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

Most accidents occurred during fine weather conditions.
